```python
# 結巴
import jieba

def chinese_word_cut(mytext):
    return " ".join(jieba.cut(mytext))
```

**train_data：所要劃分的樣本特徵集
**train_target：所要劃分的樣本結果
**test_size：樣本佔比，如果是整數的話就是樣本的數量
**random_state：是隨機數的種子。 
其實就是該組隨機數的編號，在需要重複試驗的時候，保證得到一組一樣的隨機數。比如你每次都填1，其他引數一樣的情況下你得到的隨機陣列是一樣的。但填0或不填，每次都會不一樣。
隨機數的產生取決於種子，隨機數和種子之間的關係遵從以下兩個規則：
種子不同，產生不同的隨機數；種子相同，即使例項不同也產生相同的隨機數。

```python
from sklearn.model_selection import train_test_split

#讀取 df = pd.read_excel('250+C2.xlsx', encoding='gb18030')，欄位轉str
X = pd.DataFrame(df['comment'].astype(str))
y = df.star
# 當一個函數的參數存在於一個元組或者一個字典中時，
# 用來間接的調用這個函數，
# 並將元組或者字典中的參數按照順序傳遞給參數
X['cutted_comment'] = X.comment.apply(chinese_word_cut)
# 查看 X['cutted_comment']
# select = pd.DataFrame(X, columns = ["cutted_comment"])
# print(select) 
# 查看 X['cutted_comment'][0~4]
X.cutted_comment[:5]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# print(X_train)
# print(X_test)
# print(y_train)
# print(y_test)
```

```python
def tokenizer(theText):
    tokenn = {}
    for word in theText:
        token = jieba.cut(word)
        token = list(token)
        tokenn[word] = token
    return tokenn

#loc 基於行標籤和列標籤（x_label、y_label）進行索引

N_doc = df.loc[df['star']==0]
N_list = N_doc['comment']

P_doc = df.loc[df['star']==1]
P_list = P_doc['comment']

B_doc = df.loc[df['star']==2]
B_list = B_doc['comment']
 
ntoken = tokenizer(N_list)
ptoken = tokenizer(P_list)
btoken = tokenizer(B_list)

print(len(N_list))
print(len(P_list))
print(len(B_list))
```

```python
from collections import Counter 
import operator

doc = X_train['comment']        #input 0 
```

```python
import math
from math import log

def vocabulary(list):
    # set() 函數創建一個無序不重複元素集
    vocabulary = set()
    for key in list:
        
        '''
        update() 方法用於修改當前集合，可以添加新的元素或集合到當前集合中，
        如果添加的元素在集合中已存在，則該元素只會出現一次，重複的會忽略。
        '''
        vocabulary.update(list[key])
        # print(list[key])
        
    return vocabulary



tokenn = tokenizer(doc)       #input 2
voc = vocabulary(tokenn)

# print(voc)

def rf(vocabulary , a , b , c):  # ***字典可能有問題***
    so = {}
    for word in vocabulary:
        nnum = 0
        pnum = 0
        for ntweet in a:   # ntweet X
            if word in a[ntweet]:  # N_list > ntoken  
                nnum += 1
        for ptweet in b:
            if word in b[ptweet]:  # N_list > ntoken  
                pnum += 1
        for ctweet in c:
            if word in c[ctweet]:  
                pnum += 1
    
        nnum += 1                                           #防止 NSO = 0
        so[word] = log(pnum / nnum + math.exp(1))
    return so

rf1 = rf(voc , ntoken , ptoken , btoken)
print(rf1)
```

```python
from sklearn.feature_extraction import DictVectorizer
vect = DictVectorizer(sparse=False)

t2 = vect.fit_transform(rf1)
print((t2))
```

```python
for key in rf1:
    for key2,value2 in term_matrix.iteritems():
        if key == key2:
            term_matrix[key2] = value2 * rf1[key]
```
